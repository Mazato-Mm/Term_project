```python
#Sound_Sys
import pygame
from pathlib import Path
# (Sound System)
pygame.mixer.init()

def init_audio_system():
    pygame.mixer.init()

def play_bg_music(path="bg_music.mp3", loops=-1, volume=0.5):
    p = Path(path)
    if not p.exists():
        print(f" BG music not found: {path}")
        return
    pygame.mixer.music.load(str(p))
    pygame.mixer.music.set_volume(volume)
    pygame.mixer.music.play(loops=loops)
    print(f" Playing background music: {path}")

def stop_bg_music():
    pygame.mixer.music.stop()

def pause_bg_music():
    pygame.mixer.music.pause()

def unpause_bg_music():
    pygame.mixer.music.unpause()

def set_bg_volume(volume):
    volume = max(0.0, min(1.0, volume))
    pygame.mixer.music.set_volume(volume)

def play_sound(path, volume=0.8):
    p = Path(path)
    if not p.exists():
        print(f"Sound file not found: {path}")
        return
    sound = pygame.mixer.Sound(str(p))
    sound.set_volume(volume)
    sound.play()

def play_win_sound():
    play_sound("win.mp3", volume=0.8)

def play_lose_sound():
    play_sound("lose.mp3", volume=0.8)
```
